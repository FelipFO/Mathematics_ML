##**1. Normas Lp en Numpy**

La función np.linalg.norm() devuelve la norma de un vector o matriz.
Esta función puede devolver normas matriciales diferentes, dependiendo del parámetro *ord* que reciba. A continuación se encuentran algunos ejemplos:

**ord= 'none'.** Devuelve Norma de Frobenius para matrices, norma 2 para vectores por defecto.

**ord=1.** Devuelve Norma L1, la cual es la suma de los valores absolutos de los elementos.

**ord=2.** Devuelve Norma L2, también conocida como  norma Euclidiana.

**ord= np.inf.** Norma del infinito, la cual es la suma máxima de los valores absolutos de las filas para matrices, o el valor absoluto máximo para vectores.


**1.1 Ejemplo:** Calculemos la norma Lp, con $p= 3$, para el siguiente vector:
\begin{align}
 a=\begin{pmatrix}
 2 \\
 7  \\
\end{pmatrix}
 \end{align}



```python
a=np.array([2,7])

L3=np.linalg.norm(a, ord=3)


print("Norma L3:",L3)
```

##**2. Curvas de nivel**

La norma Lp en $\mathbb{R}^n$ es una función que mapea de $\mathbb{R}^n$ a $\mathbb{R}$. Es decir, recibe un vector con $n$ elementos reales, y nos devuelve un número real. Nosotros podemos inspeccionar la geometría de una norma a través de las gráficas de las curvas de nivel en $\mathbb{R}^2$. En el siguiente ejemplo graficaremos las curvas de nivel de la norma L2, es decir, la norma Euclidiana.

```python
# Crear malla de puntos en el plano 2D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
(x, y) = np.meshgrid(x, y)

# Crear una matriz de ceros como la malla anteriormente creada
z = np.zeros_like(x)


# Calcular la norma L3 que recorre cada punto de la malla con respecto al vector dado
for i in range(len(x)):
    for j in range(len(y)):
        punto = np.array([x[i][j], y[i][j]])
        #Se le asigna el valor de la Norma L3 a cada punto de la matriz de ceros
        z[i][j] = np.linalg.norm(punto, ord= 2)

# Graficar utlizando la función contourf
fig, ax = plt.subplots()
CS1 = ax.contourf(x, y, z, alpha = 0.5, levels=10)
fig.colorbar(CS1)
ax.grid()
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set(title = 'Norma L2 en $R^2$')
plt.show()
```

**Ejercicio de práctica:** Grafique las curvas de nivel de la norma L1. Intente explicar por qué las curvas de nivel lucen como rombos centrados en el origen.

```python
# Crear malla de puntos en el plano 2D
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
(x,y) = np.meshgrid(x,y)

# Crear una matriz de ceros como la malla anteriormente creada
z = np.zeros_like(x)

# Calcular la norma L3 que recorre cada punto de la malla con respecto al vector dado
for i in range(len(x)):
  for j in range(len(y)):
    punto = np.array([x[i,j],y[i,j]])
    z[i,j] = np.linalg.norm(punto,ord=1)

fig, ax = plt.subplots()
CS1 = ax.contourf(x, y, z, alpha = 0.5, levels=10)
fig.colorbar(CS1)
ax.grid()
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set(title = 'Norma L2 en $R^2$')
plt.show()
```
